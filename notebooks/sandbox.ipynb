```python
import globalise_tools.io_tools as rw

documents = rw.read_json("../data/globalise-documents.json")
len(documents)
```

```python
mapping = {}
for d in documents:
    for s in d['series']:
        for e in s:
            path=e['path']
            uuid=e['path_uuid']
            mapping[path]=uuid
rw.write_json("../path-to-uuid.json", mapping)
```
